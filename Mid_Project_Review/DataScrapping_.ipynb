```python
import pickle as pic
import re # module to search in python
import requests
from bs4 import BeautifulSoup
import time
from googlesearch import search
import warnings
warnings.filterwarnings("ignore")
import operator
```

```python
allAlphabets = "abcdefghijklmnopqrstuvwxyz"
# To generating the next URL for the disease of a particular alphabet
def generateNHP(string,letter):
  myUrl = 'https://www.nhp.gov.in/disease-a-z/' + letter
  return myUrl
def generateCDC(string,letter):
  myUrl = 'https://www.cdc.gov/diseasesconditions/az/' + letter + '.html'
  return myUrl
```

# Scrapping data set 

```python
def all_Disease(string):
  store_Disease = []
  for letter in string:
    myUrl = generateNHP(string,letter)
    time.sleep(0.5)
    #sleep time 0.5 seconds 
    getPage= requests.get(myUrl,verify=False)
    getPageContent = BeautifulSoup(getPage.content, 'html5lib')
    disease_Html_Div = getPageContent.find('div', class_='all-disease')
    #beacuse all the disease are in the div class called 'all-disease'
    disease_Html_Div = disease_Html_Div.find_all('li')
    for disease in disease_Html_Div:
      li_disease_Text = disease.get_text()
      current_Disease = li_disease_Text.strip()
      store_Disease.append(current_Disease)


  return store_Disease

```

#Storing the diseases from nhp.gov.in

```python
allDisease = []
allDisease = all_Disease(allAlphabets)
```

# Scrapping The Data from the CDC

```python
def all_Disease_CDC(string):
  store_Disease = []
  for letter in string:
    myUrl = generateCDC(string,letter)
    time.sleep(0.5)
    #sleep time 0.5 seconds 
    getPage= requests.get(myUrl,verify=False)
    getPageContent = BeautifulSoup(getPage.content, 'html5lib')
    disease_Html_Div = getPageContent.find('div', class_='az-content col')
    #beacuse all the disease are in the div class called 'all-disease'
    disease_Html_Div = disease_Html_Div.find_all('li')
    for disease in disease_Html_Div:
      li_disease_Text = disease.get_text()
      current_Disease = li_disease_Text.strip()
      store_Disease.append(current_Disease)


  return store_Disease
```

```python
allDiseaseCDC = []
allDiseaseCDC = all_Disease_CDC(allAlphabets) 
```

# Scrapping the Symptoms from the Wikipedia Page

```python
# We have to make a new dictionary to store the wikipedia symptoms related to a particular disease
unique_Symptoms = {}

```

```python
def scrapSymps(currTable,currFlag,idx):
  for tableRow in currTable.find_all("tr"):
    tableHeader_rowScope = tableRow.find("th",{"scope":"row"})
    if tableHeader_rowScope != None :
      headerofTab = tableHeader_rowScope.get_text()
      if headerofTab == "Symptoms" :
        # we are at the header by the name of Symptoms
        currDiseaseSymp = str(tableRow.find("td"))
        currDiseaseSymp = currDiseaseSymp.replace(';',',')
        currDiseaseSymp=re.sub(r'</a>','',currDiseaseSymp)
        currDiseaseSymp=re.sub(r'<a.*?>','',currDiseaseSymp) 
        myDisease = sorted_Disease [idx]
        unique_Symptoms[myDisease] = currDiseaseSymp
        currFlag = True
        break
  return currFlag






```

```python
def scrapWiki(mySearch,idx):
  for link in search(mySearch,tld="co.in",stop=1,pause=0.4):
    queryMatcher = re.search(r'wikipedia',link)
    #if we get the word wikipedia in this link 
    foundSymptom= False #boolean variabale
    checkFlag = False
    if queryMatcher != None :
      print(link)
      wikiPage = requests.get(link,verify=False)
      currPage = BeautifulSoup(wikiPage.content, 'html5lib')
      wikiTable = currPage.find("table", {"class":"infobox"})
      if wikiTable != None :
        checkFlag = scrapSymps(wikiTable,foundSymptom,idx)
    if checkFlag:
      break





```

```python
def startExtraction():
  outerLoopRun = 0
  length = len(sorted_Disease )
  while(outerLoopRun <=length-1 ):
    #Here we will write the code
    currSearch = sorted_Disease[outerLoopRun] + ' wikipedia'
    #currSearch = new_Disease[outerLoopRun]+ ' wikipedia'
    #curr disease and wikipedia search
    scrapWiki(currSearch,outerLoopRun)
    time.sleep(2)
    outerLoopRun +=1
    



```

```python
#Have to handle the duplicacy
remove_Duplicate_Keys = set()
latest_Full_Dict ={}

def filter_Vals(myDict):
  for key,value in myDict.items():
    if value in remove_Duplicate_Keys:
      continue
    else :
     #we will add it to the new dict
     latest_Full_Dict[key] = value
     remove_Duplicate_Keys.add(value)





  
```
